# 3.Data Tranformation #Dataframe of Aggregated Transactions

```python
path1 = "C:/Users/Arsha/Desktop/Guvi/Guvi Project_(2)/pulse/data/aggregated/transaction/country/india/state/"
agg_trans_list = os.listdir(path1)

Aggregated_Transactions = {'State': [], 'Year': [], 'Quarter': [], 'Transaction_type': [], 'Transaction_count': [],
            'Transaction_amount': []}
for state in agg_trans_list:
    cur_state = path1 + state + "/"
    agg_year_list = os.listdir(cur_state)
    
    for year in agg_year_list:
        cur_year = cur_state + year + "/"
        agg_file_list = os.listdir(cur_year)
        
        for file in agg_file_list:
            cur_file = cur_year + file
            data = open(cur_file, 'r')
            A = json.load(data)
            
            for i in A['data']['transactionData']:
                name = i['name']
                count = i['paymentInstruments'][0]['count']
                amount = i['paymentInstruments'][0]['amount']
                Aggregated_Transactions['Transaction_type'].append(name)
                Aggregated_Transactions['Transaction_count'].append(count)
                Aggregated_Transactions['Transaction_amount'].append(amount)
                Aggregated_Transactions['State'].append(state)
                Aggregated_Transactions['Year'].append(year)
                Aggregated_Transactions['Quarter'].append(int(file.strip('.json')))
                
df_agg_trans = pd.DataFrame(Aggregated_Transactions)
```

# Dataframe of Aggregated User

```python
path2 = "C:/Users/Arsha/Desktop/Guvi/Guvi Project_(2)/pulse/data/aggregated/user/country/india/state/"

agg_user_list = os.listdir(path2)

Aggregated_user = {'State': [], 'Year': [], 'Quarter': [], 'Brands': [], 'Count': [],
            'Percentage': []}
for state in agg_user_list:
    cur_state = path2 + state + "/"
    agg_year_list = os.listdir(cur_state)
    
    for year in agg_year_list:
        cur_year = cur_state + year + "/"
        agg_file_list = os.listdir(cur_year)

        for file in agg_file_list:
            cur_file = cur_year + file
            data = open(cur_file, 'r')
            B = json.load(data)
            try:
                for i in B["data"]["usersByDevice"]:
                    brand_name = i["brand"]
                    counts = i["count"]
                    percents = i["percentage"]
                    Aggregated_user["Brands"].append(brand_name)
                    Aggregated_user["Count"].append(counts)
                    Aggregated_user["Percentage"].append(percents)
                    Aggregated_user["State"].append(state)
                    Aggregated_user["Year"].append(year)
                    Aggregated_user["Quarter"].append(int(file.strip('.json')))
            except:
                pass
                    
df_agg_user = pd.DataFrame(Aggregated_user)
```

# Dataframe of Map Transactions

```python
path3 = "C:/Users/Arsha/Desktop/Guvi/Guvi Project_(2)/pulse/data/map/transaction/hover/country/india/state/"

map_trans_list = os.listdir(path3)

Map_Transactions = {'State': [], 'Year': [], 'Quarter': [], 'District': [], 'Count': [],
            'Amount': []}

for state in map_trans_list:
    cur_state = path3 + state + "/"
    map_year_list = os.listdir(cur_state)
    
    for year in map_year_list:
        cur_year = cur_state + year + "/"
        map_file_list = os.listdir(cur_year)
        
        for file in map_file_list:
            cur_file = cur_year + file
            data = open(cur_file, 'r')
            C = json.load(data)
            
            for i in C["data"]["hoverDataList"]:
                district = i["name"]
                count = i["metric"][0]["count"]
                amount = i["metric"][0]["amount"]
                Map_Transactions["District"].append(district)
                Map_Transactions["Count"].append(count)
                Map_Transactions["Amount"].append(amount)
                Map_Transactions['State'].append(state)
                Map_Transactions['Year'].append(year)
                Map_Transactions['Quarter'].append(int(file.strip('.json')))
                
df_map_trans = pd.DataFrame(Map_Transactions)
```

# Dataframe of Map User

```python
path4 = "C:/Users/Arsha/Desktop/Guvi/Guvi Project_(2)/pulse/data/map/user/hover/country/india/state/"

map_user_list = os.listdir(path4)

Map_User = {"State": [], "Year": [], "Quarter": [], "District": [],
            "RegisteredUser": [], "AppOpens": []}

for state in map_user_list:
    cur_state = path4 + state + "/"
    map_year_list = os.listdir(cur_state)
    
    for year in map_year_list:
        cur_year = cur_state + year + "/"
        map_file_list = os.listdir(cur_year)
        
        for file in map_file_list:
            cur_file = cur_year + file
            data = open(cur_file, 'r')
            D = json.load(data)
            
            for i in D["data"]["hoverData"].items():
                district = i[0]
                registereduser = i[1]["registeredUsers"]
                appOpens = i[1]['appOpens']
                Map_User ["District"].append(district)
                Map_User ["RegisteredUser"].append(registereduser)
                Map_User ["AppOpens"].append(appOpens)
                Map_User ['State'].append(state)
                Map_User ['Year'].append(year)
                Map_User ['Quarter'].append(int(file.strip('.json')))
                
df_map_user = pd.DataFrame(Map_User )
```

# Dataframe of Top Transactions

```python
path5 = "C:/Users/Arsha/Desktop/Guvi/Guvi Project_(2)/pulse/data/top/transaction/country/india/state/"

top_trans_list = os.listdir(path5)

Top_Transactions = {'State': [], 'Year': [], 'Quarter': [], 'Pincode': [], 'Transaction_count': [],
            'Transaction_amount': []}

for state in top_trans_list:
    cur_state = path5 + state + "/"
    top_year_list = os.listdir(cur_state)
    
    for year in top_year_list:
        cur_year = cur_state + year + "/"
        top_file_list = os.listdir(cur_year)
        
        for file in top_file_list:
            cur_file = cur_year + file
            data = open(cur_file, 'r')
            E = json.load(data)
            
            for i in E['data']['pincodes']:
                name = i['entityName']
                count = i['metric']['count']
                amount = i['metric']['amount']
                Top_Transactions['Pincode'].append(name)
                Top_Transactions['Transaction_count'].append(count)
                Top_Transactions['Transaction_amount'].append(amount)
                Top_Transactions['State'].append(state)
                Top_Transactions['Year'].append(year)
                Top_Transactions['Quarter'].append(int(file.strip('.json')))
df_top_trans = pd.DataFrame(Top_Transactions)
```
